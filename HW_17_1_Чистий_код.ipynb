````python
import pandas as pd
import numpy as np

np.random.seed(0)
products = ["apple", "banana", "cherry", "dates", "elderberry"]
categories = ["fruit", "fruit", "fruit", "fruit", "fruit"]
N_ROWS = 10 ** 6

df = pd.DataFrame({
    "product_name": np.random.choice(products, N_ROWS),
    "product_category": np.random.choice(categories, N_ROWS),
    "product_price": np.random.uniform(1, 10, N_ROWS),
    "transaction_date": pd.date_range(start="2000-01-01", periods=N_ROWS, freq="h"),
})

mem_before = df.memory_usage(deep=True).sum()

df_opt = df.copy()
df_opt["product_name"] = df_opt["product_name"].astype("category")
df_opt["product_category"] = df_opt["product_category"].astype("category")

mem_after = df_opt.memory_usage(deep=True).sum()

ratio = mem_before / mem_after

print(f"Memory before: {mem_before / 1_048_576:.2f} MiB")
print(f"Memory after : {mem_after / 1_048_576:.2f} MiB")
print(f"↓ Reduction  : {ratio:.2f}× smaller")
````

````python
from typing import Iterable
import builtins


def calculate_variance(numbers: Iterable[float]) -> float:
    """Return the population variance of a sequence of numbers.

    Використовуємо builtins.list, щоб уникнути конфлікту, якщо десь у ноутбуці
    створили змінну з ім'ям `list` (типова помилка в Colab).
    Формула: Var(X) = sum((x - mean)**2) / n  (генеральна дисперсія).
    """
    nums = builtins.list(numbers)
    n = len(nums)
    if n == 0:
        raise ValueError("numbers must not be empty")

    mean_value = sum(nums) / n
    return sum((x - mean_value) ** 2 for x in nums) / n


# Демонстрація: очікуємо 50
input_list = [75, 80, 85, 90, 95]
result = calculate_variance(input_list)
print(result)          # 50.0
assert abs(result - 50) < 1e-12
````
